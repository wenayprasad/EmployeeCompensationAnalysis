```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "C:/Users/y/Downloads/dataset1.xlsx"
df = pd.read_excel(file_path)
df.columns = df.columns.str.strip()
df.head()
```

```python
df['Project Completion Rate'] = df['Projects Completed'] / df['Targetted Projects']
df.head()
```

```python
avg_salary_dept = df.groupby('Department')['Salary'].mean()
avg_bonus_dept = df.groupby('Department')['Bonus'].mean()
avg_completion_rate = df.groupby('Department')['Project Completion Rate'].mean()
joining_bonus_trend = df.groupby('Year')['Joining Bonus'].mean()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(8,5))
sns.barplot(x=avg_salary_dept.index, y=avg_salary_dept.values, palette="Blues_d")
plt.title("Average Salary by Department")
plt.ylabel("Average Salary")
plt.xlabel("Department")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=avg_bonus_dept.index, y=avg_bonus_dept.values, palette="Greens_d")
plt.title("Average Bonus by Department")
plt.ylabel("Average Bonus")
plt.xlabel("Department")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=avg_completion_rate.index, y=avg_completion_rate.values, palette="Oranges_d")
plt.title("Project Completion Rate by Department")
plt.ylabel("Completion Rate")
plt.xlabel("Department")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x=joining_bonus_trend.index, y=joining_bonus_trend.values, marker='o', color='purple')
plt.title("Average Joining Bonus Trend Over Years")
plt.ylabel("Average Joining Bonus")
plt.xlabel("Year")
plt.tight_layout()
plt.show()
```
